## Plotting with Matplotlib

```python
n = np.arange(0, 1000, 1)
phi = 0.5*(1+np.sqrt(5))
r = 0.5*np.sqrt(n)
theta = 2*np.pi*n/(1+phi)

x = r*np.cos(theta)
y = r*np.sin(theta)

plt.plot(x[::2],y[::2],'o')
```

```python
plt.polar(theta[::2], r[::2], 'o')
```

```python
for i in range(1,5):
    plt.polar(theta[i::5], r[i::5], 'o')
```

## Loops and control

```python
x = np.linspace(-2, 0.6, 100)
y = np.linspace(-1.2, 1.2, 100)

result = np.zeros((100, 100))
for i, xval in enumerate(x):
    for j, yval in enumerate(y):
        iters = 0
        real = 0
        imaginary = 0
        while (np.sqrt(real**2 + imaginary**2) <= 2) and iters < 70:
            real = real**2 - imaginary**2 + xval
            imaginary = 2*real*imaginary + yval
            iters +=1
        if np.sqrt(real**2 + imaginary**2) > 2:
            result[i,j] = 1 - np.sqrt(iters/70)
plt.imshow(result)
```
